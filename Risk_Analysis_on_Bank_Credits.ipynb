```python
def merged_lineplot(x,y):
    plt.figure(figsize=(8,4))
    sns.pointplot(x=x, 
                  y=y, 
                  hue="TARGET", 
                  data=merge_data,
                  palette =['g','r'])
```

```python
merged_lineplot("NAME_CONTRACT_STATUS",'DEF_60_CNT_SOCIAL_CIRCLE')
```

#### INSIGHTS: The applicants having DEF_60_CNT_SOCIAL_CIRCLE more than average of 0.13 are tends to defaults more. So, analysing social circle will help in disbursement of loan.

```python
merged_lineplot("NAME_CONTRACT_STATUS",'AMT_INCOME_TOTAL')
```

#### INSIGHTS: The line plot shows that people who has total income more than the average get defaulted because they didn't used the offer.

   * Banks must target more on contract type ‘Student’ ,’Pensioner’ and ‘Businessman’ for profitable business and housing ‘type other than ‘Co-op apartment’ for successful payments..

   * Banks must focus less on income type ‘Working’ as it is has most number of unsuccessful payments in order to get rid of financial loss for the organization.
